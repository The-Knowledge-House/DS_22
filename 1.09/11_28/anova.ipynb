# ANOVA

This notebook will introduce you to ANOVA.

```python
from scipy.stats import f

# Degrees of freedom
df_between = 2
df_within = 9

# Significance level after Bonferroni correction
alpha_bonf = 0.05 / 3  # Assuming 3 groups (k = 3)

# Critical F-value
critical_f_value = f.ppf(1 - alpha_bonf, df_between, df_within)

# Display the critical F-value
print(f"Critical F-value: {critical_f_value}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Generate x values for the F-distribution
x = np.linspace(0, 5, 1000)

# Degrees of freedom
df_between = 2
df_within = 9

# Significance level after Bonferroni correction
alpha_bonf = 0.05

# Plot the F-distribution
plt.plot(x, f.pdf(x, df_between, df_within), label=f'F-distribution (df_between={df_between}, df_within={df_within})')

# Calculate critical F-value
critical_f_value = f.ppf(1 - alpha_bonf, df_between, df_within)

# Draw line for critical F-value
plt.axvline(critical_f_value, color='red', linestyle='dashed', linewidth=1, label=f'Critical F (alpha={alpha_bonf})')

plt.title('F-distribution with Critical F-value')
plt.xlabel('F-value')
plt.ylabel('Probability Density Function')
plt.legend()
plt.grid(True)
plt.show()
```

This shows the different distributions for varying degrees of freedom between groups (more groups)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Generate x values
x = np.linspace(0, 5, 1000)

# Varying degrees of freedom
df_between_values = [1, 3, 5, 10, 50]
df_within = 10

# Plot the F-distribution for varying degrees of freedom
plt.figure(figsize=(10, 6))

for df_between in df_between_values:
    y = f.pdf(x, df_between, df_within)
    plt.plot(x, y, label=f'df_between={df_between}')

# Set y-axis limit
plt.ylim(0, 2)

plt.title('F-distribution for Varying Degrees of Freedom')
plt.xlabel('F-value')
plt.ylabel('Probability Density Function')
plt.legend()
plt.grid(True)
plt.show()
```

This shows the critical F-values that vary with increasing groups but the same sample sizes. We are also correcting the Alpha value for increasing number of groups.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Generate x values
x = np.linspace(0, 5, 1000)

# Varying degrees of freedom
df_between_values = [1, 3, 5, 10, 50]
df_within = 10

# Define colors for each critical F-value
colors = ['red', 'blue', 'green', 'purple', 'orange', 'brown']

# Plot the F-distribution for varying degrees of freedom
plt.figure(figsize=(10, 6))

for df_between, color in zip(df_between_values, colors):
    y = f.pdf(x, df_between, df_within)
    plt.plot(x, y, label=f'df_between={df_between}')

    # Significance level after Bonferroni correction
    alpha_bonf = 0.05/df_between

    # Calculate critical F-value
    critical_f_value = f.ppf(1 - alpha_bonf, df_between, df_within)
    
    # Draw line for critical F-value with different colors
    plt.axvline(critical_f_value, color=color, linestyle='dashed', linewidth=1, label=f'Critical F (df={df_between})')

plt.title('F-distribution with Critical F-values')
plt.xlabel('F-value')
plt.ylabel('Probability Density Function')
plt.ylim(0, 2)
plt.legend()
plt.grid(True)
plt.show()
```
